```python
import librosa
import sys
import matplotlib.pyplot as plt

sys.path.append('..')

from config import DATA_PATH
from pathlib import Path
from collections import Counter
from utils import get_main_freq_traj, get_spectrogram 

DRYAD_PATH = Path(DATA_PATH) / "dryad"
TRAIN_PATH = Path(DATA_PATH) / "train"
TEST_PATH = Path(DATA_PATH) / "test"
```

```python
test_usv = TRAIN_PATH/"DBA_9863_1758.wav"

# running the function from utils.py
times, freq_traj = get_main_freq_traj(test_usv)

# plot it
plt.figure(figsize=(10, 4))
plt.plot(times * 1000, freq_traj / 1000)  # convert to ms and kHz for readability
plt.xlabel("Time (ms)")
plt.ylabel("Frequency (kHz)")
plt.title(f"Main Frequency Trajectory: {test_usv.name}")
plt.show()
```

```python
times, freqs, magnitude = get_spectrogram(test_usv)

plt.figure(figsize=(10, 4))
plt.pcolormesh(times * 1000, freqs / 1000, librosa.amplitude_to_db(magnitude), shading='auto')
plt.xlabel("Time (ms)")
plt.ylabel("Frequency (kHz)")
plt.colorbar(label="dB")
plt.title(f"Spectrogram: {test_usv.name}")
plt.show()
```
